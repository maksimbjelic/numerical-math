# 3 #

```python
f = lambda x: math.e ** x - 3 * x - 1
```

```python
i = itr.Bisection(f)
print(i.run([-2, 1], 40))
i.plot([-2, 1], 5, (-1, 1))
```

```python
l = -1 / 3
i = itr.Iteration(lambda x: x - l * f(x))
print(math.log10(3))
print(i.run(1, 20))
i.plot(1, 10, (-0.5, 1.1))
i.plot_speed(1, 20, 0)
```

```python
l = -1 / 2
i = itr.Iteration(lambda x: x - l * f(x))
i.plot(1, 5, (-0.2, 1.1))
i.plot_speed(1, 5, 0)
```

```python
fig, ax = plt.subplots()

t = np.linspace(0, 2 * math.pi, 1000)
x = np.linspace(-2, 2, 1000)
ax.plot(2 * np.cos(t), np.sin(t), label='$S_1$')
ax.plot(x, math.sqrt(3) / 2 * x ** 2, label='$S_2$')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_aspect('equal')
ax.grid('on')
ax.legend()
   
plt.show()
```

```python
f1 = lambda x, y: x ** 2 + 4 * y ** 2 - 4
f2 = lambda x, y: 2 * y - math.sqrt(3) * x ** 2

J = lambda x, y: np.array([[2 * x, 8 * y], [-2 * math.sqrt(3), 2]])
```

```python
i = itr.Newton2D(f1, f2, J)
```

```python
i.run(itr.vec(2, 3), 5)
```

```python
i.plot(itr.vec(2, 3), 10, intervalx=(0.9, 2.1), intervaly=(0.8, 3.1), ans=itr.vec(1, math.sqrt(3) / 2))
```

```python
i.run(itr.vec(-1.5, 2), 5)
```

```python
i.plot(itr.vec(-1.5, 2), 5, intervalx=(-1200, 100), intervaly=(-100, 1700), ans=itr.vec(-1, math.sqrt(3) / 2))
```
